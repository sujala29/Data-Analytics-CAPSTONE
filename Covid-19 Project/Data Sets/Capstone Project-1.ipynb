# Linear Regression:  The goal is to predict the number of confirmed COVID-19 cases based on the number of total vaccinations in each state.

```python
from sklearn.linear_model import LinearRegression
import statsmodels.api as sm
```

```python
# Select the relevant columns
X = covid_df_filtered[['total_vaccinations']]
y = covid_df_filtered['Confirmed_cases']
```

```python
print(covid_df_filtered.columns)

```

```python
# Fit the linear regression model
model = LinearRegression()
model.fit(X, y)
```

```python
# Print the model coefficients
print('Coefficients:', model.coef_)
print('Intercept:', model.intercept_)
```

```python
# Perform statistical analysis
X = sm.add_constant(X) # Add a constant term to the model
model_sm = sm.OLS(y, X).fit() # Fit the model using Ordinary Least Squares
```

```python

# Print the summary of the model
print(model_sm.summary())
```
